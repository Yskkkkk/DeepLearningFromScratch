## 4.3.2 数值微分的例子

```python
import numpy as np
import matplotlib.pylab as plt

def numerical_diff(f, x):
    h = 1e-4
    return (f(x+h) - f(x-h)) / (2*h)

def function_1(x):
    return 0.01*x**2 + 0.1*x

def tangent_line(f, x):
    d = numerical_diff(f, x)
    y = f(x) - d*x
    return lambda t: d*t + y

x = np.arange(0.0, 20.0, 0.1)
y = function_1(x)

# 计算斜率
slope_5 = numerical_diff(function_1, 5)
slope_10 = numerical_diff(function_1, 10)

# 打印斜率
print(f"在x=5处的切线斜率: {slope_5:.6f}")
print(f"在x=10处的切线斜率: {slope_10:.6f}")

plt.figure(figsize=(10, 4))

# 左图：x=5
plt.subplot(1, 2, 1)
tf_5 = tangent_line(function_1, 5)
y2_5 = tf_5(x)
plt.plot(x, y, label="f(x)")
plt.plot(x, y2_5, label="tangent at x=5")
plt.title("x=5")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()

# 右图：x=10
plt.subplot(1, 2, 2)
tf_10 = tangent_line(function_1, 10)
y2_10 = tf_10(x)
plt.plot(x, y, label="f(x)")
plt.plot(x, y2_10, label="tangent at x=10")
plt.title("x=10")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()

plt.tight_layout()
plt.show()
```

## 4.3.3 偏导数

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 定义函数
def function_2(x0, x1):
    return x0**2 + x1**2

# 创建网格
x0 = np.arange(-3.0, 3.0, 0.1)
x1 = np.arange(-3.0, 3.0, 0.1)
X0, X1 = np.meshgrid(x0, x1)

# 计算函数值
Y = function_2(X0, X1)

# 绘制3D表面图
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(1, 1, 1, projection='3d')
ax.plot_surface(X0, X1, Y, cmap='viridis', alpha=0.8)
ax.set_xlabel('x0')
ax.set_ylabel('x1')
ax.set_zlabel('f(x0, x1)')
ax.set_title('f(x0, x1) = x0² + x1²')

plt.show()
```
